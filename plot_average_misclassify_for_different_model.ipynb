```python
minlog, maxlog, point = -20, 0, 21
bins = np.linspace(minlog, maxlog, point)
cbin = 0.5 * (bins[1:] + bins[:-1])
```

```python
plt.plot(cbin, zero_arr, label=r'$ \alpha = 0.0$ (Regular Train)')
# plt.plot(cbin, one_arr)
plt.plot(cbin, two_arr, label=r'$ \alpha = 0.01 $')
# plt.plot(cbin, three_arr)
# plt.plot(cbin, four_arr, label=r'$ \alpha = 0.05 $')
plt.plot(cbin, five_arr, label=r'$ \alpha = 0.1 $')
# plt.plot(cbin, six_arr)
plt.plot(cbin, seven_arr, label=r'$ \alpha = 0.3 $')
plt.plot(cbin, eight_arr, label=r'$ \alpha = 0.5 $')
plt.plot(cbin, nine_arr, label=r'$ \alpha = 1.0 $')
plt.ylim([0, 30])
plt.legend()
plt.xlabel('Curvature (Log Scale)')
plt.ylabel('Average Misclassify Samples')
plt.title(r' Average Misclassify Samples with Different $\alpha$ under same $\sigma$ = 0.2')
```
